# 00 Method Smoke Test

## Purpose

Run a minimal synthetic workflow that creates exceedances, estimates a runs extremal index, computes return times, and saves one figure and one table.

## Data / config used

- synthetic logistic-map observable
- output directory: `outputs/notebooks/00_method_smoke_test/`


```python
from pathlib import Path

import numpy as np
import pandas as pd

from dynsys_econometrics.extremes import runs_extremal_index, threshold_exceedances
from dynsys_econometrics.plots import plot_series_with_threshold
from dynsys_econometrics.return_times import return_times
from dynsys_econometrics.simulation import simulate_logistic_map

output_dir = Path("outputs/notebooks/00_method_smoke_test")
output_dir.mkdir(parents=True, exist_ok=True)
```

```python
panel = simulate_logistic_map(n=1500, r=4.0, x0=0.17, burn_in=100)
panel["value"] = -np.log(np.abs(panel["value"] - 0.5) + 1e-12)
events = threshold_exceedances(panel[["date", "series_id", "value"]], quantile=0.95)
summary = runs_extremal_index(events, run_length=4)
durations = return_times(events)

events.to_csv(output_dir / "events.csv", index=False)
summary.to_csv(output_dir / "extremal_index.csv", index=False)
durations.to_csv(output_dir / "return_times.csv", index=False)

threshold = float(events["threshold"].iloc[0])
series = panel.set_index("date")["value"]
plot_series_with_threshold(series, threshold=threshold, output_path=output_dir / "smoke_threshold.png")

summary
```

## Limitations

This notebook is synthetic and methodological. It does not claim that an observed economy is a deterministic map, and the diagnostics are descriptive rather than causal.
